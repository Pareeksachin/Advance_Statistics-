```python
"""
Covariance : it measures the linear realtionship between two variables. It indicates how changes in one variable are associated with changes in another variable
 Covariance helps identify the direction and strength of the relationship between variables but does not provide a standardized measure.
 
Correlation : 
Correlation measures the strength and direction of the linear relationship between two variables. It is a standardized measure that ranges from -1 to 1. 
A correlation of +1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 
0 indicates no linear relationship. In the Iris dataset, we can calculate the correlation between features such as sepal length and sepal width.
If we find a high positive correlation between these two features, it suggests that as the sepal length increases, the sepal width also tends to 
increase consistently. Correlation helps quantify the strength and direction of the relationship between variables, allowing us to make more precise 
interpretations compared to covariance.
"""
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 56, 78, 90, 109, 123])
y = np.array([789, 896, 908, 1023, 1348, 1789])
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
mean_x
```

```python
mean_y
```

```python
dev_x = x-mean_x
dev_x
```

```python
dev_y = y - mean_y
dev_y
```

```python
dev_x*dev_y
```

```python
covariance = np.mean(dev_x*dev_y)
covariance

# now we don't understand the direct relation through this, to know the relation we need to calculate correlation coefficient
```

```python
# in one line 
covariance = np.mean((x-np.mean(x))*(y-np.mean(y)))
covariance
```

```python
# to calculate standard deviation 
#Step 1: Calculate the mean
# Step 2: Calculate the squared difference
# Step 3: Calculate the mean of squared differences
# Step 4: Take the square root to get the standard deviation
squred_diff = (x-mean_x)**2
squred_diff
```

```python
mean_squared_diff = np.mean(squred_diff)
mean_squared_diff
```

```python
std_deviation = np.sqrt(mean_squared_diff)
std_deviation

# YOU CAN DO SAME FOR y 
```

```python
# Standard Deviation WITH HELP OF INBUILT FUNCTION
std_x = np.std(x)
std_y = np.std(y)
```

```python
std_x
```

```python
std_y
```

```python
correlation = covariance/(std_x*std_y)
print("Correlation =", correlation)
```

```python
np.corrcoef(x, y)
print(np.cov(x,y,ddof=0))
#calculates the correlation coefficient matrix between two arrays x and y
#Both np.corrcoef and np.cov are built-in functions in NumPy that provide convenient ways to calculate the correlation coefficient and covariance matrix, respectively. 
```
