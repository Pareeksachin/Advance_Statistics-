```python
import pandas as pd 
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
iris = pd.read_csv(r"C:\Users\Anonymous\Desktop\Advance Statistics\Datasets\iris.csv")
```

```python
iris
```

```python
# Before performing PCA on a dataset

# -> Data preparation : Insure that your dataset is ready for PCA. It involves Handling missing values, Dealing with categorical variable(if any)
# --- standardizing the numberic features,PCA is sensitive to the scale of the variables, so it's important to standardize them to 
# have zero mean and unit variance.

# Feature Selection: If you have a large number of features, you might want to consider feature selection techniques to reduce the dimensionality 
# ----of the dataset

```

```python
# Seperate the features(X) and the target variable(Y)
X = iris.iloc[:,:-1]
Y = iris.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
# Standardize the fatures
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#The process of standardizing the features involves transforming the data such that it has zero mean and unit variance. This is done to ensure that all
# ----features are on the same scale and have similar ranges.

# By standardizing the features, now have a mean of 0 and a standard deviation of 1. This ensures that both
#    features contribute equally to the analysis and avoids any issues caused by differences in their scales.
```

```python
X_scaled
```

```python
"""If we don't standardize the features before performing PCA, the variables with larger scales will dominate the analysis
and have a higher influence on the principal components. This can lead to biased results and misinterpretation of the relative 
importance of each feature.

In the case of the iris dataset, if we don't standardize the features, the variables with larger scales, such as "Sepal.Length"
and "Petal.Length," will have a greater impact on the principal components compared to the variables with smaller scales, such
as "Sepal.Width" and "Petal.Width."

For example, the "Sepal.Length" has a range of values from 4.3 to 7.9, while the "Sepal.Width" has a range from 2.0 to 4.4.
Without standardization, the "Sepal.Length" values will contribute more to the principal components due to their larger 
magnitude, even if "Sepal.Width" contains valuable information.

Standardizing the features ensures that each feature has a mean of 0 and a standard deviation of 1. This puts all the features 
on a similar scale, allowing them to contribute equally to the PCA analysis. It removes the bias introduced by the original
scales and allows for a fair comparison of the features' importance in explaining the variability of the dataset."""
```

```python
#Perform PCA 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
"""we first load the iris dataset and separate the features (X) and the target variable (y). 
Then, we standardize the features using StandardScaler to ensure they are on the same scale.
Next, we perform PCA using PCA from the sklearn.decomposition module, specifying the number of components as 2.

We transform the standardized features into the principal components using the fit_transform method. The resulting
principal components are stored in X_pca. We create a new DataFrame pca_df to combine the principal components and 
the target variable."""
```

```python
# create a new dataframe with the principle componenet and the target variable

pca_df = pd.DataFrame(data = X_pca,columns = ['Principal Component 1','Principal Component 2'])
pca_df['Species'] = Y
```

```python
pca_df
```

```python
# Explained variance ratio

#A higher variance indicates that the data points are more spread out and have larger differences from the mean,
# while a lower variance indicates that the data points are closer to the mean and have smaller differences.
#In linear regression, variance is used to estimate the accuracy of the regression model. In hypothesis testing,
# variance is used to assess the variability within different groups or populations.
#in PCA, variance is used to determine the importance of each principal componen
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio :",explained_variance_ratio)

#In the context of Principal Component Analysis (PCA), the explained variance ratio represents the proportion of 
 #the dataset's variance explained by each principal component. It indicates how much information or variability
# is captured by each principal component.

# A higher explained variance ratio for a principal component suggests that it captures a larger portion of the dataset's variance. 
```

```python
# plot the principal componenets 
plt.figure(figsize = (10,6))
sns.scatterplot(data=pca_df,x = 'Principal Component 1', y = 'Principal Component 2',hue = 'Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - IRIS Dataset')
plt.show()
```
