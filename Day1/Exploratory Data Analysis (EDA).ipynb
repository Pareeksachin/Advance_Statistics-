```python
counts = cars['AirBags'].value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=False)
counts
#The bins parameter in the value_counts() function is used when you want to divide a numerical Series into 
# ---equal-width bins and count the occurrences in each bin

# bins argument only work with numeric data
```

```python
# plot a bar chart 
plt.bar(counts.index,counts)
plt.show()
# you will get error : TypeError: 'value' must be an instance of str or bytes, not a float
# Here matplotlib library used for plotting , expects a the values to be of type 'str' or 'bytes' for 
# --- categorical data. However the 'NaN' value is considered as a flot, causing the TypeError
# To resolve this issue you can convert 'NaN' vvalue to a string representation before creating the bar plot
```

```python
counts = cars['AirBags'].value_counts(dropna = False)
# Convert NaN to string representation
counts = counts.fillna('NaN')
# Convert index to strings
counts.index = counts.index.astype(str) #bcz index of the counts Series is of type float
plt.bar(counts.index,counts)

plt.show()
```
