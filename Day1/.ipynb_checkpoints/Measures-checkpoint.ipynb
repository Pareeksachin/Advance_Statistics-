```python
import numpy as np 
a = np.array([20,89,0,29,102,58,8,79])
b = np.array([43,28,28,29,37,38,28,33,35])
```

```python
print("Mean of A : ",np.mean(a))
print("Mean of B : ",np.mean(b))
```

```python
rng = np.max(a)-np.max(b)
print("Range = ",rng)
```

```python
q1 = np.quantile(a,0.25)
q3 = np.quantile(a,0.75)
qd = (q3-q1)/2
print("Quartile Deviation = ",qd) 
# The quartile deviation is a measure of dispersion that provides an estimate of the spread of data around the median
```

```python
# Mean Absolute Deviation (MAD) 
#MAD is a measure of dispersion that quantifies the average absolute difference between each data point and the mean of the dataset.
mad_a = np.mean(np.absolute(a-np.mean(a)))
print("Mean Deviation = ",mad_a)
# It is less affected by outliers compared to other measures like standard deviation
```

```python
# variance 
var_a = np.mean((a-np.mean(a))**2)
print("Variance = ",var_a)
print("Variance with help of function = ",np.var(a))
#variance should always be a non-negative value. If you are getting a negative value for 
# --- the variance, it suggests there might be an issue with the data or the calculations.
```

```python
print("Mean = ",np.mean(b))
rng = np.max(b)-np.min(b)
print("Range = ",rng)
q1 = np.quantile(b,0.25)
q3 = np.quantile(b,0.75)
qd = (q3-q1)/2
print("Quartile Deviation = ",qd)
mad_b = np.mean(np.absolute(b-np.mean(b)))
print("Mean Deviation = ",mad_b)
var_b = np.mean((b-np.mean(b))**2)
print("Variance = ",var_b)
np.var(b)
```

```python
# Function 
def stats(x):
    print("Mean = ",np.mean(x))
    print("variance = ",np.var(x))
    print("Std Deviation = ",np.std(x))
    print("Coeff of Variation = ",(np.std(x)/np.mean(x))*100)
    print("------------------------------------------")
    
weight = np.array([65,78,45,90,73,40])
stats(weight)

height = np.array([169,170,132,100,137,155])
stats(height)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars = pd.read_csv("C:/Users/Anonymous/Desktop/Advance Statistics/Datasets/Cars93.csv")
cars['Price'].plot(kind = 'box')
plt.show()
```

```python
q1 = np.quantile(cars['Price'], 0.25)
q3 = np.quantile(cars['Price'], 0.75)
iqr = q3 - q1 
print("Inter-Quartile Range =", iqr)
print("1.5 IQR =", 1.5*iqr)
```

```python
cars93 = pd.read_csv("C:/Users/Anonymous/Desktop/Advance Statistics/Datasets/Cars93.csv")
upper_fence = cars93[cars93['Price'] > 1.5*iqr + q3 ]
upper_fence['Price']

lower_fence = cars93[cars93['Price'] < q1 - 1.5*iqr ]
lower_fence['Price']
```

```python
def box_outliers(column, dataframe):  
    q1 = np.quantile(dataframe[column].dropna(), 0.25)
    q3 = np.quantile(dataframe[column].dropna(), 0.75)
    iqr = q3 - q1 
    upper = dataframe[dataframe[column] > 1.5*iqr + q3 ]
    print("Upper Outliers :")
    print(list(upper[column]))
    
    lower = dataframe[dataframe[column] < q1 - 1.5*iqr ]
    print("Lower Outliers =") 
    print(list(lower[column]))
    
    plt.boxplot(dataframe[column].dropna())
    plt.title("Boxplot of " + column)
    plt.show()

    
box_outliers('Price', cars93)    
box_outliers('Weight', cars93)

cars93['MPG.highway'].plot(kind='box')
plt.show()

box_outliers('MPG.highway', cars93)
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import ipywidgets as widgets
```

```python
cars = pd.read_csv("C:/Users/Anonymous/Desktop/Advance Statistics/Datasets/Cars93.csv")
cars.head()
```

```python
cars93['AirBags'].fillna('NaN', inplace=True)
sns.boxplot(x='AirBags',y='Price',data = cars93)
plt.show()
```

```python
g = sns.FacetGrid(cars93,col = 'AirBags')
g = g.map(plt.hist,"Price",bins = 5)
plt.show()
```

```python
cars['Price'].skew()
```

```python
cars['Price'].kurtosis()
```

```python
plt.hist(cars['Price'])
plt.show()
```

```python
def boxplot(var,category):
    cars['AirBags'].fillna('NaN', inplace=True)
    sns.boxplot(x = category,y = var,data = cars)
    plt.show()
```

```python
boxplot(var = 'Price',category = 'AirBags')
```

```python
# using Widgets 
widgets.interact(boxplot,var = ["Price","MPG.city","MPG.highway","Horsepower","EngineSize"],
                category = ["AirBags","Origin","DriverTrain","Man.trans.avail"])
```

```python
def histogram(var,bins):
    plt.hist(var,bins,data = cars)
    plt.show()
```

```python
widgets.interact(histogram,var=["Price","MPG.city","MPG.highway","Horsepower","EngineSize"],
                bins=(5,25,5))
```

```python
def groupby_bar(var, category):
    cts = cars.groupby(category)[var].mean()
    plt.bar(cts.index,cts)
    plt.show()
groupby_bar('Price', 'AirBags')
```

```python
widgets.interact(groupby_bar, var=["Price","MPG.city","MPG.highway","Horsepower","EngineSize"], 
                category=["AirBags","Origin","DriveTrain","Man.trans.avail","Type"])
```

```python
# Pie Chart
```

```python
cts = cars['AirBags'].value_counts()
plt.pie(cts,labels=cts.index,autopct='%1.1f%%')
plt.show()
```

```python
def pie(category):
    cts = cars[category].value_counts()
    plt.pie(cts,labels=cts.index,autopct='%1.1f%%')
    plt.show()
```

```python
widgets.interact(pie, category=["AirBags","Origin","DriveTrain","Man.trans.avail","Type"])
```
